# Q4 Replace all the 0s in the column with the median of the same column value accordingly.

```python
# Replace 0 with median
df1['Plas'] = df1['Plas'].replace(0, df1['Plas'].median())
df1['Pres'] = df1['Pres'].replace(0, df1['Pres'].median())
df1['skin'] = df1['skin'].replace(0, df1['skin'].median())

```

# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
X=df1.drop(labels='class',axis=1) # droping column cloumn "class"
y=df1['class'] # assigning column "class"
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1) # splitting into 70 and 30
```

# Q9 Create the decision tree model using â€œentropyâ€ method of reducing the entropy and fit it to training data.</b>



```python
dt_model = DecisionTreeClassifier(criterion = 'entropy' )
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_test , y_test)
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
y_predict = dt_model.predict(X_test)
print(metrics.confusion_matrix(y_test, y_predict))
```

```python
dt_model.score(X_test, y_test)
```
