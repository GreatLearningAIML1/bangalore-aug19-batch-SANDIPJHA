### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
X=df1.drop(labels='default',axis=1) # Features are assigned to X
y=df1['default'] #target is assigned to y
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
# as per given number there will be 30% for test set and 70% for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999)
```

```python
rfcl.fit(X_train,y_train) # fit model
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
print("Test Score :",rfcl.score(X_test,y_test)) # Test Score
```

```python
print("Training Score :",rfcl.score(X_train,y_train)) # Training score
```

```python
from sklearn import metrics
y_predict=rfcl.predict(X_test)
print(metrics.confusion_matrix(y_test, y_predict)) # confusion matrix
```

```python
print("Test F1:Score: ", metrics.f1_score(y_test, y_predict))  # Test F1 score
```
